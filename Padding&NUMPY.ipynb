```python
import numpy as np
```

```python
x=np.arange(1,19).reshape(3,6)
print(x)
```

```python
y=np.pad(x,pad_width=1,mode="constant",constant_values=5)
print(y)
```

```python
y=np.pad(x,pad_width=1,mode="edge")
print(y)
```

```python
y=np.pad(x,pad_width=1,mode="wrap")  #wrap is opposite
print(y)
```

```python
y=np.pad(x,pad_width=1,mode="reflect")  #it will reflect like that hide the first column and reflect second column
print(y)
```

```python
y=np.pad  (x,pad_width=1,mode="symmetric")# it will show the first row
print(y)
```

```python
y=np.pad  (x,pad_width=2,mode="symmetric")
print(y)
```

```python
y=np.pad(x,pad_width=1,mode='minimum')
y
```

```python
y=np.pad(x,pad_width=2,mode='minimum')
y
```

```python
y=np.pad(x,pad_width=1,mode='maximum')
y
```

```python
y=np.pad(x,pad_width=1,mode='mean')
y
```

```python
y=np.pad(x,pad_width=1,mode='median')
y
```

```python
y=np.pad(x,pad_width=5,mode="constant",constant_values=5)
print(y)
```

```python
y=np.pad(x,pad_width=2,mode="empty")
print(y)
# this no advisible dont use
```

```python
y=np.pad(x,pad_width=1,mode="linear_ramp")
y
```
